```python
import pandas as pd
import re
import tqdm
df = pd.read_csv('BankTranscripts.csv')
df
```

```python
'''
regex encontrar cantidades de dinero

'''
```

```python
tqdm.tqdm.pandas()
def money(transcript):
    money_regex = r'[\$€¥]\d+[\d,.]*'
    matches = re.findall(money_regex, transcript)
    if matches:
        matches = [match.replace(',', '') for match in matches]
        return matches
    else:
        return None 
```

```python
df['money_transcript'] = df['Transcript'].progress_apply(money)
```

```python
df
```

```python
def email(transcript):
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    matches = re.findall(email_regex, transcript)
    if matches:
        return matches
    else:
        return None
```

```python
df['email_transcript'] = df['Transcript'].progress_apply(email)
```
