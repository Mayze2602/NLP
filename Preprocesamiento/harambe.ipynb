```python
import pandas as pd
df = pd.read_csv('RIPHarambe.csv')
df.head()
```

```python
sentence = ' '.join(df['Tweet']) 
sentence
```

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

stopwords_en = stopwords.words('english')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import contractions
sentence = contractions.fix(sentence)
sentence = sentence.lower()
sentence = re.sub(r'_','', sentence) #Quitar guion bajo
sentence = re.sub(r'@[a-z0-9]+', ' ', sentence) #Quitar menciones
sentence = re.sub(r'\\[n]', ' ', sentence) #Quitar saltos de linea
sentence = re.sub(r'https:\/\/t\.co\/[a-z0-9]{10}', ' ', sentence) #Quitar URLs de este tipo https
sentence = re.sub(r'[^0-9a-z]', ' ', sentence) #Quitar caracteres especiales
sentence = re.sub(r'x[0-9a-z]{2}','',sentence) #Quitar emojis
sentence = re.sub(r'\s+', ' ', sentence) #Quitar espacios extras
sentence = re.sub(r'b rt', '', sentence) #Quitar rt
sentence = re.sub(r'\bb\b', '', sentence) #Quitar palabras de una sola letra

sentence
```

```python
sentence_tokenized = word_tokenize(sentence)
sentence_tokenized
```

```python
sentence_no_stopwords = [token for token in sentence_tokenized if token not in stopwords_en]
sentence_no_stopwords
```

```python
final = [lemmatizer.lemmatize(token) for token in sentence_no_stopwords]
final
```

```python
final = ' '.join(final)
final
```

```python
from PIL import Image
import numpy as np
mask = np.array(Image.open('harambe6.jpeg'))
mask.shape
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width = 1200, 
                      height = 1200,
                      colormap='viridis',
                      mask=mask,
                      background_color='white',
                      min_font_size=4).generate(final)

plt.figure(figsize=(5, 5))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout()
plt.show()
```
